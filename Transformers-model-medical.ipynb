```python
import gradio as gr
from huggingface_hub import InferenceClient

model = InferenceClient(base_url='https://api-inference.huggingface.co/models/meta-llama/Llama-3.2-3B-Instruct')

messages = [{"role": "system", "content": "You are an experienced medical AI assistant designed to provide evidence-based information on diseases, treatments, and medical research, while emphasizing patient safety and ethical considerations. Refuse questions that aren't related to medicine."}]

def chat_with_memory_and_context(user_question, context=None):
    global messages
    if context:
        messages.append({"role": "system", "content": context})
    messages.append({"role": "user", "content": user_question})
    
    for m in messages:
        m["content"] = str(m["content"])
    if len(messages) > 3:
        print("Message à l'index 3 :", messages[3])

    try:
        response = model.chat_completion(messages, max_tokens=1024)
    except Exception as e:
        print("Erreur lors de l'appel à l'API :", e)
        raise

    assistant_reply = response.get('choices', [{}])[0].get('message', {}).get('content', "I couldn't generate a response.")
    messages.append({"role": "assistant", "content": assistant_reply})
    return assistant_reply

def chatbot(question, context=None):
    return chat_with_memory_and_context(question, context)

demo = gr.ChatInterface(fn=chatbot)
demo.launch()
```
